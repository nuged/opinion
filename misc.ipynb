```python
import pandas as pd

df = pd.read_csv('mydata/merged/vaccine.tsv', index_col=None, sep='\t', quoting=3)
```

```python
def foo(row):
    bounds = eval(row.bounds)
    absence = eval(row.absence)
    result = []
    for a, b in zip(absence, bounds):
        if a == 1:
            start = b[0] - 15 if b[0] > 15 else 0
            end = b[1] + 5 if b[1] + 5 < len(row.text) else len(row.text)
            result.append(row.text[start: end])
    if result:
        return result
```

```python
for t in df.apply(foo, axis=1).tolist():
    if t:
        print(t)
```
